Run the following load command to get the 2-Dim iris dataset. Do the following operations on it.

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data'
iris_2d = np.genfromtxt(url, delimiter=',', dtype='float', usecols=[0,1,2,3])
```

Drop rows that contain a missing value. 

```python
print(print(iris_2d[~np.isnan(iris_2d).any(axis=1)]))
```

Find the correlation between first two columns.

```python
irish1=iris_2d[:,:1]
irish2=iris_2d[:,1:2]
print(np.corrcoef(irish1,irish2,rowvar=False))
```

Find the mean,median and standard deviation of column 3.

```python
irish3 = iris_2d[:,2]
print(np.average(irish3), np.median(irish3), np.std(irish3))
```

Create a new column for volume in iris_2d, where volume is (pi x petallength x sepal_length^2)/3

```python
arr_pl = iris_2d[:,2]
arr_sl = (iris_2d[:,0] ** 2)
arr_vol = (3.14*arr_pl*arr_sl)/3
arr_vol=arr_vol.reshape(150,1)
final=np.concatenate((iris_2d,arr_vol),axis=1)
print(final)
```

Bin the petal length (3rd) column of iris_2d to form a text array, such that if petal length is:
Less than 3 –> ‘small’
3-5 –> ‘medium’
‘>=5 –> ‘large’

```python
Bin = np.array([])
for i in iris:
  if(float(i[2])<3):
    Bin = np.append(Bin, "small")
  elif(float(i[2])>3 and float(i[2])<5):  
    Bin = np.append(Bin, "medium")
  else:
    Bin = np.append(Bin, "large")
print(Bin)
```
